Sort a dictionary by values

```python
# Original dictionary
marks = {
    "Math": 88,
    "English": 75,
    "Science": 93,
    "History": 80
}

print("Original Dictionary:")
print(marks)

# Sort by values (ascending)
sorted_by_value_asc = dict(sorted(marks.items(), key=lambda item: item[1]))
print("\nSorted by values (ascending):")
print(sorted_by_value_asc)

# Sort by values (descending)
sorted_by_value_desc = dict(sorted(marks.items(), key=lambda item: item[1], reverse=True))
print("\nSorted by values (descending):")
print(sorted_by_value_desc)

# Sort by keys (alphabetically)
sorted_by_key = dict(sorted(marks.items()))
print("\nSorted by keys (alphabetically):")
print(sorted_by_key)
```
